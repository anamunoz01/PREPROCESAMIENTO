```python
import cv2 as cv
```

```python
from google.colab.patches import cv2_imshow
imagen = cv.imread("monedas.png")
cv2_imshow(imagen)
```

```python
imagen_gray = cv.cvtColor(imagen, cv.COLOR_BGR2GRAY)
cv2_imshow(imagen_gray)
```

```python
import numpy as np
```

```python
ret, imagenbw = cv.threshold(imagen_gray,127, 255, cv.THRESH_BINARY)
cv2_imshow(imagenbw)
```

```python
Brush = np.ones((5,5),np.uint8)
Brush1 = np.ones((8,8),np.uint8) #elemento estructurante
```

erosionamos la imagen

```python
image_erode = cv.erode(imagenbw, Brush, iterations = 1)
cv2_imshow(image_erode)
```

Dilatamos la imagen

```python
image_dilate = cv.dilate(imagenbw, Brush1, iterations = 1)
cv2_imshow(image_dilate)
```

abrir imagen = erosionar y dilatar

```python
image_open = cv.morphologyEx(imagenbw, cv.MORPH_OPEN, Brush)
cv2_imshow(image_open)
```

cerrar imagen = dilatar y erosionar

```python
image_close = cv.morphologyEx(imagenbw, cv.MORPH_CLOSE, Brush1)
cv2_imshow(image_close)
```

Gradiante

• Es la diferencia entre la dilatación y la erosión de una 
imagen. El resultado se parecerá al contorno del 
objeto.

```python
image_gradient = cv.morphologyEx(imagenbw, cv.MORPH_GRADIENT, Brush)
cv2_imshow(image_gradient)
```

Top HAT

• la diferencia entre la imagen de entrada y la apertura 
de la imagen

```python
image_tophat = cv.morphologyEx(imagenbw, cv.MORPH_TOPHAT, Brush)
cv2_imshow(image_tophat)
```

Back HAT

• Es la diferencia entre el cierre de la imagen y la 
imagen de entrada.

```python
image_blackhat = cv.morphologyEx(imagenbw, cv.MORPH_BLACKHAT, Brush)
cv2_imshow(image_blackhat)
```
